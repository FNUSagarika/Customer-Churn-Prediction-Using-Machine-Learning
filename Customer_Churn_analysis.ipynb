## 3.1.Histograms of 'Gender' and 'Geography'


```python
import plotly.express as px

def hist(feature):
    group_df = data.groupby([feature, 'Exited']).size().reset_index()
    group_df = group_df.rename(columns={0: 'Count'})
    fig = px.histogram(group_df, x=feature, y='Count', color='Exited', marginal='box', title=f'Churn rate frequency to {feature} distribution', color_discrete_sequence=["green", "red"])
    fig.show()
```

```python
hist('Geography')
hist('Gender')
```

```python
# Churn rates for Geography
churn_rate_country = data.groupby("Geography")["Exited"].mean()
churn_rate_country.index = churn_rate_country.index.map({0: "France", 1: "Germany", 2: "Spain"})
print("Churn rate by country:\n", churn_rate_country)
```

```python
print(churn_rate_country.idxmax(), 'has higher customer churn rate')
```

```python
# Churn rates for Gender
churn_rate_gender = data.groupby("Gender")["Exited"].mean()
churn_rate_gender.index = churn_rate_gender.index.map({0: "Female", 1: "Male"})
print("\nChurn rate by gender:\n", churn_rate_gender)
```

```python
print(churn_rate_gender.idxmax(), 'customer more likely to stop the credit card ')
```

```python
import plotly.express as px
import pandas as pd
from scipy import stats
# Function to perform t-test and provide insight into the relationship between feature and churn
def analyze_feature_for_churn(feature):
    churned = data[data["Exited"] == 1][feature]
    non_churned = data[data["Exited"] == 0][feature]

    # Perform t-test to see if there is a significant difference between churned and non-churned customers
    t_stat, p_val = stats.ttest_ind(churned, non_churned)

    # Print results
    print(f"Analysis for {feature}:")
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_val}")

    if p_val < 0.05:
        print(f"The difference in {feature} is statistically significant with p-value {p_val}.")
    else:
        print(f"The difference in {feature} is not statistically significant with p-value {p_val}.\n")

# 1. Analyze Credit Score for churn
analyze_feature_for_churn('CreditScore')

# 2. Analyze Estimated Salary for churn
analyze_feature_for_churn('EstimatedSalary')

```

## 4.Predictive data analysis

```python
X = data[["CreditScore", "Geography", "Gender", "Age", "Tenure", "EstimatedSalary"]]
y=data[["Exited"]]
```

```python
# split train and test data
from sklearn.model_selection import train_test_split
Train_X, Test_X, Train_y, Test_y = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
Train_X
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier().fit(Train_X.values, Train_y)
```

```python
prediction = dtc.predict(np.array([[445,0,1,64,2,43678.06],[640,0,0,20,4,78310.82]]))
print('Prediction:', prediction)
```

```python
from matplotlib import pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(25,20))
tree1 = tree.plot_tree(dtc,
                   feature_names=Train_X.columns.values,
                   class_names=['No','Yes'],
                   filled=True)
```

```python
tree2 = tree.plot_tree(dtc,
                   max_depth=2,
                   feature_names=Train_X.columns.values,
                   class_names=['No','Yes'],
                   filled=True)
```

Predicitons on the test set.

```python
Test_y_pred = dtc.predict(Test_X)
Test_y_pred, Test_y
```

```python
from sklearn.metrics import classification_report
print(classification_report(Test_y, Test_y_pred))
```

```python
from matplotlib import pyplot
importance = dtc.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.title("Feature Importance")
pyplot.xticks([x for x in range(len(importance))], Train_X.columns)
pyplot.tight_layout()
pyplot.show()
```
